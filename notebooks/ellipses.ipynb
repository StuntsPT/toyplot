```python
import numpy
numpy.random.seed(1234)

import toyplot
```

```python
rx = 2
ry = 1
angle = 0

canvas = toyplot.Canvas()
axes = canvas.cartesian(xmin=-3, xmax=3, ymin=-2, ymax=2)
axes.ellipse(0, 0, rx, ry, angle, style={"stroke":"black", "fill":"none"})

theta = numpy.deg2rad(angle)
axes.scatterplot(rx * numpy.cos(theta), rx * numpy.sin(theta))
axes.scatterplot(ry * numpy.cos(theta + numpy.pi / 2), ry * numpy.sin(theta + numpy.pi / 2));
```

```python
mean = numpy.array([0, 0])
cov = numpy.array([[1, 0], [0, .2]])
samples = numpy.random.multivariate_normal(mean=mean, cov=cov, size=1000)

sigx = cov[0, 0]
sigy = cov[1, 1]

p = 0.95
s = -2 * numpy.log(1 - p)
print(s)

canvas = toyplot.Canvas()
axes = canvas.cartesian(xmin=-3, xmax=3, ymin=-3, ymax=3)
#axes = canvas.cartesian()

axes.scatterplot(samples[:,0], samples[:,1]);
axes.ellipse(0, 0, sigx * numpy.sqrt(s), sigy * numpy.sqrt(s), 0, style={"stroke":"black", "fill":"none"})
#axes.ellipse(0, 0, cov[0, 0] * 2, cov[1, 1] * 2, 0, style={"stroke":"black", "fill":"none"})
#axes.ellipse(0, 0, cov[0, 0] * 3, cov[1, 1] * 3, 0, style={"stroke":"black", "fill":"none"})
```
