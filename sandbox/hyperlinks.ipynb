```python
import toyplot
canvas = toyplot.Canvas(style={"background-color":"lightgray"}, hyperlink="#canvas")
axes = canvas.cartesian(hyperlink="#axes")
axes.text(0, 0, "<a href='#text'>Click Me!</a>");
```

```python
canvas = toyplot.Canvas()
axes = canvas.cartesian()
axes.bars([1, 2, 3], hyperlink=["#1", "#2", "#3"]);
```

```python
canvas = toyplot.Canvas()
axes = canvas.cartesian()
axes.bars([[1, 2], [3, 4], [5, 6]], hyperlink=["#1", "#2", "#3"], baseline=None);
```
