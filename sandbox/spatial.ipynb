```python
import logging
import numpy
import toyplot
toyplot.log.setLevel(logging.DEBUG)
```

```python
numpy.random.seed(1234)
x = numpy.linspace(-10, 10, 500)
y = numpy.sin(x)
coordinates = numpy.column_stack((x, y))
```

```python
import toyplot.spatial
qtree = toyplot.spatial.QuadTree(bbox=(
    coordinates[:,0].min(),
    coordinates[:,0].max(),
    coordinates[:,1].min(),
    coordinates[:,1].max(),
    ))
for index, coordinate in enumerate(coordinates):
    qtree.insert(item=index, point=coordinate)
    
print qtree
```

```python
area = (-0.49, 0.49, -0.49, 0.49)
intersection = qtree.intersect(area)
print len(intersection)
```

```python
color = numpy.repeat("black", len(coordinates)).astype("S10")
color[list(intersection)] = "red"

canvas = toyplot.Canvas()
axes = canvas.cartesian(aspect=None)
axes.scatterplot(coordinates[:,0], coordinates[:,1], color=color)
axes.rects(area[0], area[1], area[2], area[3], style={"fill":"none", "stroke":"red"})
def walk(node):
    axes.rects(node.bbox.xmin, node.bbox.xmax, node.bbox.ymin, node.bbox.ymax, style={"fill":"none"})
    for child in node.nodes:
        walk(child)
walk(qtree.root)
```
