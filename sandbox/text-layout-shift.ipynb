```python
import toyplot.font
import toyplot.text
```

```python
text = "Test <small>me!</small>"
style = {
    "font-size": "16px",
    "-toyplot-text-layout-visibility": "visible",
    "-toyplot-text-layout-line-visibility": "visible",
    "-toyplot-text-layout-box-visibility": "visible",
}
fonts = toyplot.font.ReportlabLibrary()
```

```python
canvas = toyplot.Canvas(width=300, height=150)
axes = canvas.cartesian()
mark = axes.text(0, 0, text, style=style)

layout = toyplot.text.layout(text, style, fonts)
toyplot.text.dump(layout)

print "\nextents:", mark.extents(["x", "y"])[1]
```
